```python
# Load dataset in and view the first five rows
busData = pd.read_csv("cleaneddata_sql.csv")
busData.head()
```

## Delete unecessary columns

```python
def latitude(bloc):
    # long = float(bloc.split('(')[1].split(' ')[0])
    lat = float(bloc.split('(')[1].split(' ')[1].split(')')[0])
    return(lat)
```

```python
def longitude(bloc):
    long = float(bloc.split('(')[1].split(' ')[0])
    #lat = float(bloc.split('(')[1].split(' ')[1].split(')')[0])
    return(long)
```

```python
clean_data = busData
```

```python
# Loop
clean_data = clean_data[clean_data['Business Location'].apply(type) != float]
clean_data['lat'] = clean_data['Business Location'].apply(latitude)
clean_data['long'] = clean_data['Business Location'].apply(longitude)
```

```python
# Save the cleaned data into PostgreSQL

from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:postgres@localhost:5432/businessreg')
clean_data.to_sql('sf_business_1', engine) #sf_business_1 is a new table, which will be created by PostgreSQL
```
